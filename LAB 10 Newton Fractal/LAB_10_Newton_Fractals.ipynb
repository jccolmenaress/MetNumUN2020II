##Newton iterates form z=10.0 + 0.0j

```
z=10.0 + 0.0j
r=[]
i=[]
for k in range(50):
  dz = df(z)
  z = z - f(z) / dz # Newton iteration
  r.append(np.real(z))
  i.append(np.imag(z))
plt.title('Newton iterates form z=10.0 + 0.0j')
plt.plot(r)
plt.ylabel('Real part')
plt.show()
plt.plot(i)
plt.ylabel('Imagnary part')
plt.show()
```

Con un z=10.0 + 0.0j la funcin $z^{6}+z^{3}-1$ en su componente real converge en un valor de 1, y en su componente complejo converge en 0.
